```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import matplotlib.font_manager as fm
myfont1=fm.FontProperties(fname='C:/Windows/Fonts/times.ttf',size=30)

myfont2=fm.FontProperties(fname='C:/Windows/Fonts/times.ttf',size=15)

#读取数据文件
a=np.loadtxt("Experimental data.txt")

#数据分类
t=a[:,0]
OD600 = [a[:,1],a[:,2],a[:,3]]
MCLR = [a[:,4],a[:,5],a[:,6]]

#数据处理
#np.mean(a,axis=1) 每列取均值
#np.mean(a,axis=0)每行取均值
OD600_mean=np.mean(OD600, axis=0)
MCLR_mean=np.mean(MCLR, axis=0)

OD600_std = np.std(OD600,axis=0)
MCLR_std=np.std(MCLR,axis=0)

fig=plt.figure()
plt.figure(num=3, figsize=(8, 5))

fig,ax1=plt.subplots()
#ax1.plot(t,OD600_mean,'s-',label='line one',color='g')
#设置次坐标轴
ax2=ax1.twinx()
#ax2.plot(t,MCLR_mean,'*-',label='line one')

#画误差线
ax1.errorbar(t,OD600_mean, yerr=OD600_std, fmt='s-',color='red',linewidth=3)
ax2.errorbar(t,MCLR_mean, yerr=MCLR_std, fmt='-o',color='black',linewidth=3)

#font=FontProperties(fname=r"/Library/Fonts/times.ttc",size=16)

#设置x和y轴标签
ax1.set_ylabel('A',fontproperties=myfont1)
ax2.set_ylabel('MCLR (mg/L)',fontproperties=myfont1)
ax1.set_xlabel('Time (h)',fontproperties=myfont1)
ax1.tick_params(direction='in',width=2,length=4,size=5)
ax2.tick_params(direction='in',width=2,length=4,size=5)

#设置纵坐标范围
ax1.set_ylim(-0.01,0.14)
ax2.set_ylim(-0.5,7)
ax1.set_xlim(-0.5,16)
#plt.yticks(fontsize=20)
#plt.xticks(fontsize=20)

ax1.legend(['OD600'],loc=(.70,.90),numpoints=1,fancybox=False,fontsize=25)

ax2.legend(['MCLR'],loc=(.70,.82),numpoints=1,fancybox=False,fontsize=25)

plt.show()

fig.set_size_inches(12, 10)
fig.savefig('degrade.png', dpi=400,bbox_inches='tight')
```
